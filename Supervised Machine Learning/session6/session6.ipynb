
1. Download a dataset of mobile phone prices (with features like RAM, storage, camera MP, battery, brand, etc.) and fit a multiple linear regression model to predict the price using at least three features of your choice.


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

mobile = pd.DataFrame({
    "RAM": [4, 6, 8, 12, 16, 8, 6, 12],
    "Storage": [64, 128, 128, 256, 512, 256, 64, 512],
    "Camera_MP": [48, 64, 64, 108, 200, 108, 50, 200],
    "Battery": [4000, 4500, 5000, 5000, 6000, 5500, 4500, 6000],
    "Price": [15000, 22000, 28000, 42000, 70000, 38000, 21000, 68000]
})

X = mobile[["RAM", "Storage", "Camera_MP"]]
y = mobile["Price"]

model = LinearRegression()
model.fit(X, y)

print("Model Trained Successfully")
```

2. Display the regression coefficients and intercept from your model, and write a short comment on what each coefficient means in the context of mobile phones (e.g., how much price increases per extra GB of RAM).


```python
print("Intercept:", model.intercept_)

coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

print(coefficients)
```

3. Using the statsmodels or scikit-learn library, calculate the Variance Inflation Factor (VIF) for each feature in your dataset and identify if any features show signs of multicollinearity.<br><br><em><strong>Hint:</strong> VIF values above 5 or 10 often indicate problematic multicollinearity.</em>


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print(vif)
```

4. Interpret the feature importances or coefficients from your regression model to decide which feature is most influential in predicting mobile phone price, and explain your reasoning in 2-3 sentences.


```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

print(importance.sort_values(by="Coefficient", ascending=False))
```

5. Refactor your code to remove one highly collinear feature (based on VIF) and re-fit the regression model. Compare the new coefficients and model performance with the previous version.<br><br><em><strong>Hint:</strong> Removing multicollinear features can make your model more stable and interpretable.</em>


```python
X_new = mobile[["RAM", "Camera_MP"]]

new_model = LinearRegression()
new_model.fit(X_new, y)

print("New Intercept:", new_model.intercept_)

new_coefficients = pd.DataFrame({
    "Feature": X_new.columns,
    "Coefficient": new_model.coef_
})

print(new_coefficients)

old_score = model.score(X, y)
new_score = new_model.score(X_new, y)

print("\nOld R² Score:", old_score)
print("New R² Score:", new_score)
```
